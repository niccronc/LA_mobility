```python
import time
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
def get_ridership(line):
    '''
    This function outputs a list containing the ridership data for a specific bus line.
    Each element of the list is the data for a specific month, in string format
    '''
    time.sleep(3)
    
    css_selector_line='select#ContentPlaceHolder1_lbLines > option[value="'+line+'"]'
    driver.find_element_by_css_selector(css_selector_line).click()

    driver.find_element_by_css_selector('input[name="ctl00$ContentPlaceHolder1$btnSubmit"]').click()
    
    time.sleep(3)
    
    try:
        table=driver.find_element_by_css_selector('table#ContentPlaceHolder1_ASPxRoundPanel2_GridView1')
        return table.text.split('\n')[1:]
    
    except NoSuchElementException:
        return []
```
