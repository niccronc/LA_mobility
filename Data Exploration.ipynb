```python
df['date'] = pd.to_datetime(df.start_time)
```

```python
def extract_day(row):
    row['date'] = row['date'].date()
    return row
```

```python
df = df.apply(extract_day, axis=1)
```

```python
bike_trips_by_day = df.groupby('date').count().trip_id
```

```python
bike_trips_by_day.plot()
```
